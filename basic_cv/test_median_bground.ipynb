```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

url ='https://github.com/secutron/TesTime/raw/main/data/sample_cctv.mp4'
cap = cv2.VideoCapture(url)

#  50 
random_Ids = cap.get(cv2.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=25)
random_frames = []
for id in random_Ids:
    cap.set(cv2.CAP_PROP_POS_FRAMES, id)
    ret, frame = cap.read()
    random_frames.append(frame)

# median  
median_frame = np.median(random_frames, axis=0).astype(dtype=np.uint8)   

cv2_imshow(median_frame)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
